```python
import cv2  # Import OpenCV library for image processing
import numpy as np  # Import NumPy library for numerical operations
import matplotlib.pyplot as plt  # Import Matplotlib library for visualization
from ultralytics import YOLO  # Import YOLO object detection model from Ultralytics
import easyocr  # Import EasyOCR library for optical character recognition
```

# **Load model**



```python
model = YOLO("yolov8s.yaml")  # build a new model from scratch
```

```python
# Use the model
model.train(data="/content/drive/MyDrive/config.yaml", epochs=80,batch=32,imgsz=640,save=True)  # train the model
```

```python
# to get the model evaluation
metrics = model.val()
```

```python
# Load class list from file
with open("/content/drive/MyDrive/mydata.txt", "r") as file:
    class_list = file.read().split("\n")
```

```python
# Initialize YOLO model
model = YOLO('/content/best.pt')
```

```python
# Read the image
frame = cv2.imread('/content/drive/MyDrive/Data_car/images/test/1429.jpg')
frame = cv2.resize(frame, (640, 640))
frame_copy = frame.copy()
```

```python
# Perform object detection
results = model.predict(frame)
bounding_boxes = results[0].boxes.data
```

```python
# Process detected objects
objects = []
cropped_images = []
for box in bounding_boxes:
    x1, y1, x2, y2, _, class_index = map(int, box)
    class_name = class_list[class_index]
    # Crop the area within the bounding box
    cropped_image = frame[y1:y2, x1:x2]

    cx, cy = (x1 + x2) // 2, (y1 + y2) // 2

    if 'car name plate' in class_name:
        color = (255, 0, 0)  # Red for person with mask
    else:
        color = (0, 0, 255)  # Blue for other classes

    objects.append((cx, cy, color))
    cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)

# Convert the image to RGB (OpenCV uses BGR by default)
frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

# Display the image with bounding boxes using Matplotlib
plt.figure(figsize=(10, 10))
plt.imshow(frame_rgb)
plt.axis('off')  # Turn off axis labels
plt.show()
```

```python
cropped_images = {}  # Dictionary to store cropped images by class name

for box in bounding_boxes:
    x1, y1, x2, y2, _, class_index = map(int, box)
    class_name = class_list[class_index]

    # Crop the area within the bounding box
    cropped_image = frame[y1:y2, x1:x2]

    if class_name not in cropped_images:
        cropped_images[class_name] = []  # Initialize list if class not seen before

    cropped_images[class_name].append(cropped_image)

# Display the cropped images
plt.figure(figsize=(10, 10))
for i, (class_name, images) in enumerate(cropped_images.items(), start=1):
    for j, cropped_image in enumerate(images, start=1):
        plt.subplot(len(cropped_images), len(images), i * j)
        plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
        plt.title(f"{class_name}_{j}")
        plt.axis('off')

plt.show()
```

```python
reader = easyocr.Reader(['ar'])  # Initialize EasyOCR with Arabic as the language

# Loop through cropped images and perform OCR
for class_name, images in cropped_images.items():
    for j, cropped_image in enumerate(images, start=1):
        results = reader.readtext(cropped_image)

        # Iterate over OCR results for each cropped image
        for detection in results:
            print(detection[1])  # Detected text
```
